```python
import pandas as pd
```

```python
dict_infos = {
    'ages': [10, 13, 15, 19, 20, 22, 44, 45, 67, 67],
    'height': [1.6, 1.63, 1.69, 1.72, 1.72, 1.78, 1.82, 1.83, 1.85, 1.85]
}
dataframe_infos = pd.DataFrame.from_dict(dict_infos)
```

### HISTOGRAMA

```python
dataframe_infos['ages'].hist()
```

### BARRAS

```python
dict_vendas = {'categorias': ['Esportes', 'Casual', 'Formal'],
                    'faturamento': [350000, 620000, 200000],
                    'quantidade': [1200, 4000, 2300]}
dataframe_vendas = pd.DataFrame.from_dict(dict_vendas)
```

```python
# BARRAS VERTICAL
dataframe_vendas.plot.bar(x='categorias', y='faturamento')
```

```python
# BARRAS VERTICAL ORDENADA - RANKING
dataframe_vendas.sort_values('faturamento', ascending=False).plot.bar(x='categorias', y='faturamento')
```

```python
# BARRAS HORIZONTAL ORDENADA - RANKING
dataframe_vendas.sort_values('faturamento').plot.barh(x='categorias', y='faturamento')
```

```python
# GRAFICO DE DISPERSO
dataframe_infos.plot.scatter(x='ages', y='height')
```

```python
# GRFICO DE BOXPLOT - IDADES
dataframe_infos['ages'].plot.box()
```

```python
# GRFICO DE BOXPLOT - ALTURAS
dataframe_infos['height'].plot.box()
```

### LINHAS

```python
dict_vendas_date = {'date_ref': ['2021-01-01', '2022-01-01', '2023-01-01', '2024-01-01', '2025-01-01'],
                    'faturamento': [8000, 80000, 37000, 200000, 102000]}
dataframe_vendas_date = pd.DataFrame.from_dict(dict_vendas_date)
dataframe_vendas_date['date_ref'] = pd.to_datetime(dataframe_vendas_date['date_ref'])
```

```python
dataframe_vendas_date.plot.line(x='date_ref', y='faturamento')
```
